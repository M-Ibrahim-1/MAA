# Deleting Columns with NaN

```python
# Calculate the percentage of NaN values in each column
nan_percentages = (df_0.isnull().sum() / len(df_0)) * 100

# Define the threshold for the percentage of NaN values
threshold = 1

# Get the columns that exceed the threshold and store them in a list
columns_to_drop = nan_percentages[nan_percentages > threshold].index.tolist()

# Drop the columns that have more than 1 percent NaN values
df_0.drop(columns=columns_to_drop, inplace=True)
df_0
```

<hr>

# Deleting constant values

```python
def delete_columns_with_high_zeros(df, value, threshold=1):
    # Calculate the percentage of zero values in each column
    zero_percentages = (df.eq(value).sum() / len(df)) * 100

    # Get the columns that exceed the threshold and store them in a list
    columns_to_drop = zero_percentages[zero_percentages > threshold].index.tolist()

    # Drop the columns that have more than 1 percent zeros
    df.drop(columns=columns_to_drop, inplace=True)

    return df
# Call the function to delete columns with more than 1 percent zeros
result_df = delete_columns_with_high_zeros(df_0, value= 0)
# Call the function to delete columns with more than 1 percent ones
result_df = delete_columns_with_high_zeros(df_0, value= 1)

# Print the resulting DataFrame
result_df
```

# Deleting certain columns by names

```python
# Specify the prefix of the columns you want to delete
substrings_to_delete = ['time.', 'Unreal', 'Unnamed', 'RdBus.L1.TyreScaling']
# RdBus.L1.TyreScaling has the same values as time
for substring in substrings_to_delete:
    # Filter columns that contain the specified substring
    filtered_columns = df_0.filter(like=substring).columns

    # Drop the filtered columns from the DataFrame
    df_0.drop(columns=filtered_columns, inplace=True)
df_0
```

# Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named 'df'

# Determine the number of columns and groups of 10 columns
num_columns = len(df_0.columns)
num_groups = num_columns // 5 + 1 if num_columns % 5 != 0 else num_columns // 5

# Create subplots to hold box plots for each group of 10 columns
fig, axes = plt.subplots(num_groups, figsize=(20, 80))

# Iterate through each group of 10 columns and plot the box plots
for i in range(num_groups):
    start_col = i * 5
    end_col = (i + 1) * 5
    group_columns = df_0.columns[start_col:end_col]
    
    # Box plot for the current group of 10 columns
    df_0[group_columns].boxplot(ax=axes[i])
    
    # Set title and labels for the plot
    axes[i].set_title(f"Box Plots for Columns {start_col+1} to {end_col}")
    axes[i].set_xlabel("Columns")
    axes[i].set_ylabel("Values")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
